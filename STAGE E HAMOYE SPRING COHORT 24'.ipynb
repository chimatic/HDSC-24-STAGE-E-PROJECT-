```python
#HERE IS A FUNCTION TO CLEAN THE DATA, REMOVE NULL VALUES, WRANGLE AND RETURN AN UPDATED DATAFRAME

def new(df):
    #replace NaN in index 0
    df.fillna('Consolidated Statements of Income', inplace = True)
    
    #set column names equal to values in row index position 0
    df.columns = df.iloc[0]
    
    #remove first row from dataframe
    df = df[1:]
    
    #reset index values
    df.reset_index(drop=True, inplace=True)
    
    #view updated dataframe
    return df
```

#### We are considering year 2023, year 2020 and year 2017 to get a seven year data span from year 2017 to year 2023

```python
final_df23 = new(df23)
```

```python
final_df20 = new(df20)
```

```python
final_df19 = new(df19)
```

```python
#Let us change the column header names to be uniform
update_name = {'Dec. 31, 2023':'2023','Dec. 31, 2022':'2022','Dec. 31, 2021':'2021','Dec. 31, 2020':'2020',
              'Dec. 31, 2019':'2019','Dec. 31, 2018':'2018','Dec. 31, 2017':'2017','Dec. 31, 2016':'2016'}
```
